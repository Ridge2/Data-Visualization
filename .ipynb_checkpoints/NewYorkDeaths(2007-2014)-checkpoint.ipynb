```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
%matplotlib inline
dataSet = pd.read_csv("https://data.cityofnewyork.us/resource/uvxr-2jwn.csv")
```

```python
dataSet.head()
```

```python
dataSet['death_rate'] = dataSet['death_rate'].replace('NaN', np.nan)
dataSet['age_adjusted_death_rate'] = dataSet['age_adjusted_death_rate'].replace('NaN', np.nan)
dataSet['deaths'] = dataSet['deaths'].replace('NaN', np.nan)
```

```python
dataSet.head()
```

```python
dataSet = dataSet.dropna()
dataSet
```

```python
axes = {'x': ['sex', 'deaths'], 'y': [dataSet['sex'], dataSet['deaths']]}
sns.barplot(y = axes['y'], x = axes['x'], orient='v');
```
